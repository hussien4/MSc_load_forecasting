```python
df = pd.read_csv('/avg consumption per hour and month.csv')
```

```python
df['avg(hour)'].nunique()
```

```python
df['avg(hour)'] = df['avg(hour)'].astype('int')
```

```python
df['hour'] = pd.Categorical(df['avg(hour)'])
```

```python
color_used = sb.color_palette()[0]
plt.figure(figsize=[8,2.5])
sb.boxplot(y = 'avg(energy)', x= 'hour',data =df ,color = color_used)

plt.xticks(fontsize = 11.5,rotation =90)
plt.yticks(fontsize = 11.5)

plt.xlabel('Hour' ,fontsize =16)
plt.ylabel('Avg. energy (kWh)' ,fontsize =13)
from google.colab import files
plt.savefig("Box plot hour vs avg energy.png",dpi = 300 , bbox_inches = 'tight')
files.download("Box plot hour vs avg energy.png")
```
