```python
# no of houses for each acorn category
df1 = df.groupby(['stdorToU','Acorn_grouped']).count()['LCLid'].reset_index()
```

```python
i = df.loc[df['Acorn_grouped'] == 'ACORN-'].index
df = df.drop(index = i)
```

```python
u = df.loc[df['Acorn_grouped'] == 'ACORN-U'].index
df = df.drop(index = u)
```

```python
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman'] + plt.rcParams['font.serif']
```

```python
csfont = {'fontname':'Times New Roman'}
plt.figure(figsize= [4,3])
sb.countplot(data = df , x ='Acorn_grouped',hue='stdorToU'  )
plt.xlabel("Acorn group" ,fontsize = 14 )
plt.ylabel("Count of houses" ,fontsize = 14)
plt.xticks(fontsize = 12,rotation=0)
plt.yticks(fontsize = 11)

plt.legend(title=None,edgecolor = 'k' , framealpha=1)
from google.colab import files
plt.savefig("count of houses for each acorn category and for each tariff type.png",dpi=500,bbox_inches='tight')
files.download("count of houses for each acorn category and for each tariff type.png")
```
